```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# MLP-evaluation

#source = 'evaluation_k/MLP_evaluation_k_total.csv'
#source = 'evaluation_k_l/MLP_evaluation_kl_total.csv'
#source = 'evaluation_k_l_t/MLP_evaluation_k=5_t_total.csv'
source = 'evaluation_k_l_t/MLP_evaluation_k_t=0.1_total.csv'
#source = 'evaluation_k_l_t/MLP_evaluation_k_t=0.09_total.csv'


#save_name = "Graph/MLP_evaluation_k.png"
#save_name = "Graph/MLP_evaluation_kl.png"
#save_name = "Graph/MLP_evaluation_k=5_t.png"
save_name = "Graph/MLP_evaluation_k_t=0.1.png"
#save_name = "Graph/MLP_evaluation_k_t=0.09.png"

Title   = 'MLP_evaluation'
X_label = "K & T=0.1"
Y_label = ""

data = pd.read_csv(source, delimiter=",")
#print(data.head())

xdata = []
y1data = []
y2data = []

xdata = data.loc[:, 'K']
y1data = data.loc[:, 'Accuracy']
y2data = data.loc[:, 'Precision']
y3data = data.loc[:, 'Recall']
y4data = data.loc[:, 'F1_score']

fig, ty1 = plt.subplots(figsize=(10, 8))
plt.title(Title)
plt.xlabel(X_label)
plt.ylabel(Y_label)

plt.plot(xdata, y1data, marker = 'o', markersize = 5)
plt.plot(xdata, y2data, marker = 'o', markersize = 5)
plt.plot(xdata, y3data, marker = 'o', markersize = 5)
plt.plot(xdata, y4data, marker = 'o', markersize = 5)

plt.legend(['Accuracy', 'Precision', 'Recall', 'F1_score'] )

plt.savefig(save_name)
plt.show()
```
